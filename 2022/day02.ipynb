```python
with open("day02.input", "r") as f:
    seq = [tuple(t.strip().split()) for t in f.readlines()]

print(seq)
```

```python
def game(pair):
    scores = {
        ("A", "Y"): 6,
        ("A", "Z"): 0,
        ("B", "X"): 0,
        ("B", "Z"): 6,
        ("C", "X"): 6,
        ("C", "Y"): 0,
    }
    return scores.get(pair, 3)
```

```python
def score_individual(pair):
    scores = {
        "X":1,
        "Y":2,
        "Z":3
    }
    return scores[pair[1]]
```

```python
final_scores = [game(pair)+score_individual(pair) for pair in seq]
print(sum(final_scores))
```

```python
def game2(pair):
    gesture = {
        ("A", "X"): "C",
        ("A", "Y"): "A",
        ("A", "Z"): "B",
        ("B", "X"): "A",
        ("B", "Y"): "B",
        ("B", "Z"): "C",
        ("C", "X"): "B",
        ("C", "Y"): "C",
        ("C", "Z"): "A",
    }
    score_ges = {
        "A": 1,
        "B": 2,
        "C": 3
    }
    score_win = {
        "X": 0,
        "Y": 3,
        "Z": 6
    }
    return score_ges[gesture[pair]] + score_win[pair[1]]
```

```python
final_scores = [game2(pair) for pair in seq]
print(sum(final_scores))
```
